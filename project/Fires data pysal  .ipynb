## Explore data on active fires from Near Real-Time (NRT) Suomi National	Polar-orbiting	Partnership Visible Infrared Imaging Radiometer Suite (SNPP/VIIRS) Active Fire detection
---

* What I did:
    - I downloaded data from [EARTHDATA](https://earthdata.nasa.gov/earth-observation-data/near-real-time/firms/active-fire-data) in shapefile format.
    - I used [PYSAL](http://pysal.readthedocs.io/en/latest/) to read the data to the dataframe.
    - I used [Leaflet.js](http://leafletjs.com) to visualize the fires near DC region.    
        - I actually simply used [folium](https://pypi.python.org/pypi/folium) python module to create map. 



```python
#shp_path = '../VNP14IMGTDL_NRT_Global_7d/VNP14IMGTDL_NRT_Global_7d.shp'
#shp_path = '../VNP14IMGTDL_NRT_Central_America_24h/VNP14IMGTDL_NRT_Central_America_24h.shp'
shp_path = '../VNP14IMGTDL_NRT_USA_contiguous_and_Hawaii_24h/VNP14IMGTDL_NRT_USA_contiguous_and_Hawaii_24h.shp'

#USA_contiguous_and_Hawaii
fires_df = ps.pdio.read_files(shp_path)

#print(fires_df.describe())
dc_coord = [38.889931, -77.009003]
delta = 5.0
dc_fires = fires_df[(fires_df["LATITUDE"] < dc_coord[0]+delta) &
                    (fires_df["LATITUDE"] > dc_coord[0]-delta) &
                    (fires_df["LONGITUDE"] < dc_coord[1]+delta) &
                    (fires_df["LONGITUDE"] > dc_coord[1]-delta)
                   ]
#print(dc_fires.describe())
```

```python
map_fires = folium.Map(location=dc_coord,zoom_start=7,tiles='Stamen Terrain')
for i,elem in dc_fires.iterrows():
#    print(elem["geometry"])
    folium.CircleMarker([elem["LATITUDE"],elem["LONGITUDE"]],
                    radius=1000,
                    popup=str(elem['BRIGHT_TI4']),
                    color='#3186cc',
                    fill_color='#ff3314',
                   ).add_to(map_fires)
    #    folium.Marker(elem["geometry"], popup='Marker %d'%i).add_to(map_fires)
map_fires
```
